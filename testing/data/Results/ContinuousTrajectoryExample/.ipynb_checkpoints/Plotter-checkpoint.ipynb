```python
#import required modules
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import json
import itertools
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
#load file
DIR = "C:/Users/olgaa/Desktop/Sebas/Thesis/Results/ContinuousTrajectoryExample/"
ext = ".json"
```

```python
def loadFile(file):
    with open(DIR + file + ext) as json_file:
        data = json.load(json_file)
        #process data
        data["trajectory"] = np.reshape(data["trajectory"], (-1,3))
        for solver in data["solvers"]:
            solver["trajectoryGenerated"] = np.reshape(solver["trajectoryGenerated"], (-1,3))
        return data
```

```python
def graphTrajectoryBenchmark(file, ymin = None, ymax = None, step = 5, info = ""):
    solvers_stats = loadFile(file)
    N = len(solvers_stats["solvers"])
    
    fig = plt.figure(figsize=(15,10 * N))
    #graph generated trajectories
    colormap = mpl.cm.Dark2.colors

    for i, color in zip(range(N), colormap):
        ax = fig.add_subplot(2 * N, 2 , 4 * i + 1, projection='3d')
        #graph real trajectory
        real_trajectory = solvers_stats["trajectory"]
        ax.plot(real_trajectory[:,0],real_trajectory[:,1],real_trajectory[:,2], color = mpl.cm.Dark2.colors[0], label = "Real_Trajectory", marker = ".", linewidth= 2*(N + 1), alpha = 0.4) 
        solver_name = solvers_stats["solvers"][i]["name"]
        solver_trajectory = solvers_stats["solvers"][i]["trajectoryGenerated"]
        alpha = 1 #(i+1)/(N+1)
        ax.plot(solver_trajectory[:,0],solver_trajectory[:,1],solver_trajectory[:,2], color = color, label = solver_name, alpha=alpha, linewidth = 3, linestyle='--')
    
        ax.set_ylabel('Trajectory')
        ax.set_xlabel('Points')
        ax.set_title('Generated trajectory' + info)
        ax.legend(loc='lower left')
        
        #graph error of generated path vs true path
        ax = fig.add_subplot(2 * N, 2 , 4 * i + 2)
        ax.plot(solvers_stats["solvers"][i]["errorValues"], color = color, label = solver_name, marker = ".")    
        ax.set_ylabel('Distance between target and end effector')
        ax.set_xlabel('Trajectory point')
        ax.set_title('Distance between target and effector' + info)
        ax.legend(loc='lower left')
        

        #graph distance between frames
        ax = fig.add_subplot(2 * N, 2 , 4 * i + 3)
        ax.plot(solvers_stats["solvers"][i]["distanceValues"], color = color, label = solver_name, marker = ".")    
        ax.set_ylabel('Mean distance between postures')
        ax.set_xlabel('Posture')
        ax.set_title('Mean distance between postures' + info)
        ax.legend(loc='lower left')

        #graph distance between frames
        ax = fig.add_subplot(2 * N, 2 , 4 * i + 4)
        ax.plot(solvers_stats["solvers"][i]["orientationDistanceValues"], color = color, label = solver_name, marker = ".")    
        ax.set_ylabel('Mean orientation distance between postures')
        ax.set_xlabel('Posture')
        ax.set_title('Mean orientation distance between postures' + info)
        ax.legend(loc='lower left')
        
        
    plt.show()
```

```python
graphTrajectoryBenchmark("continuous_path_joints_10_postures_1000_NONE_lissajous_x_1.0_y3.0_z3.0", ymin = None, ymax = None)
```
